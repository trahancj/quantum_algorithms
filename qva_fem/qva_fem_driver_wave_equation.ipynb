```python
import numpy as np
import matplotlib.pyplot as plt
import random
from fem_1d import *
from qva import *
from pauliDecomp import *
from scipy.sparse import diags

# Check the answer and classical FE solution
nqbits = 2 #3 #2
nlayers = 3#. 6 #3
maxiter = 200 #2000 #200
reduced = 1
method = "COBYLA"
rhobeg = np.pi/10.0

# Define and Solve the FEM Problem
def uan(x,t):
    return (np.sin(x + t))
xl = 0; xr = 1; nx = 2**(nqbits) + 2;
tl = 0; tr = 1; nt = 12
ni = nx - 2
reduced = 1
c=1
u_analytic, x, t, M, A, dt, h = wave_solve(c,xl,xr,nx,tl,tr,nt,reduced,uan)

# Initial QVA Parameters
random.seed()
nparameters = nqbits + 2*(nqbits-1)*(nlayers-1)
parameters = [float(random.randint(-3100,3100))/1000 for i in range(0, nparameters)]

# Get FEM Matrix Pauli Decomposition
result = pauli_decomp(nqbits,A,Pauli,Pauli_Names,nPauli)
c = []; g = []; 
for i in range(len(result)):
    gt = [None] * nqbits
    for j in range(nqbits):
        gt[nqbits-j - 1] = result[i][2][j]
    c.append(result[i][0])
    g.append(gt)

# get initial conditions
n1 = 0
older_time = t[n1]
old_time   = t[n1]+dt
#print("older time: ",older_time);
#print("old time: ",old_time);
u_older = uan(x,older_time)
u_old = uan(x,old_time)

# Store results
u_store = []; cost_values_store = [];
u_store.append(u_older.tolist())
u_store.append(u_old.tolist())

h,x = create_grid(nx,xl,xr)
hh,xp = create_grid(101,xl,xr)
fig = plt.figure(figsize=(6, 6),facecolor=(1, 1, 1))
plt.title('time: %.1fs'%(older_time),color="black",fontsize=18)
plt.plot(x,u_older,color="black",linestyle='--',marker='o',markersize=15,markerfacecolor='none',linewidth=1.0)
plt.plot(xp,uan(xp,older_time),color="black")
plt.show()

h,x = create_grid(nx,xl,xr)
hh,xp = create_grid(101,xl,xr)
fig = plt.figure(figsize=(6, 6),facecolor=(1, 1, 1))
plt.title('time: %.1fs'%(old_time),color="black",fontsize=18)
plt.plot(x,u_old,color="black",linestyle='--',marker='o',markersize=15,markerfacecolor='none',linewidth=1.0)
plt.plot(xp,uan(xp,old_time),color="black")
plt.show()

for n1 in range(2,nt):
    older_time = t[n1]-2*dt
    old_time   = t[n1]-dt
    new_time   = t[n1]
    
    #print("older time: ",older_time);
    #print("old time: ",old_time);
    #print("new time: ",new_time);
        
    ul_older = uan(xl,older_time) # updated bc at next time
    ur_older = uan(xr,older_time) # updated bc at next time
    ul_old = uan(xl,old_time) # updated bc at next time
    ur_old = uan(xr,old_time) # updated bc at next time
    ul_new = uan(xl,new_time) # updated bc at next time
    ur_new = uan(xr,new_time) # updated bc at next time
    
    # boundary conditions and scaling values
    uscale = uan(x[1],new_time)

    # calculate RHS
    if reduced == 1:
        b = 2*np.matmul(M,u_old[1:nx-1]) - np.matmul(M,u_older[1:nx-1])
        b[0] = b[0] - A[0,1] * ul_new + (2 * M[0,1]*ul_old - M[0,1]*ul_older)
        b[ni-1] = b[ni-1]  - A[0,1] * ur_new + (2 * M[0,1]*ur_old - M[0,1]*ur_older)
#     else:
#         u_old = u_heat(x,tl)
#         b = np.matmul(M,u)
#         b[0] = ul
#         b[n-1] = ur

    #############################################################################################################
    #nit = 0
    #cost_values = []
    u, parameters, cost_values = run_qva(nqbits,nlayers,maxiter,c,g,b,parameters,method,rhobeg,ul_new,ur_new,uscale,reduced)
    #############################################################################################################

    cost_values_store.append(cost_values)
    u_store.append(u.tolist())
    
    fig = plt.figure(figsize=(6, 6),facecolor=(1, 1, 1))
    plt.title('time: %.1fs'%(new_time),color="black",fontsize=18)
    plt.plot(x,u,color="black",linestyle='--',marker='o',markersize=15,markerfacecolor='none',linewidth=1.0)
    plt.plot(xp,uan(xp,new_time),color="black")
    plt.show()
    
    u_older = u_old
    u_old = u
```

```python
from sympy import *
x, t, h, xl, xr, xm, dt = symbols('x t h xl xr xm dt')

xl = xm - h
xr = xm + h

phi1_e1 = (x-xl)/h
phi1_e2 = (xr-x)/h

phi0_e1 = (xm-x)/h
phi2_e2 = (x-xm)/h

f = diff(phi1_e1,x)*diff(phi1_e1,x)
g = diff(phi1_e2,x)*diff(phi1_e2,x)
diag = dt*dt*integrate(f,(x,xl,xm)) + dt*dt*integrate(g,(x,xm,xr))
print("K_diagonal: ",simplify(diag))

f = diff(phi1_e1,x)*diff(phi0_e1,x)
sdiag = dt*dt*integrate(f,(x,xl,xm))
print("K_subdiag above: ",simplify(sdiag))

f = diff(phi1_e2,x)*diff(phi2_e2,x)
sdiag = dt*dt*integrate(f,(x,xm,xr))
print("K_subdiag below: ",simplify(sdiag))

f = phi1_e1*phi1_e1
g = phi1_e2*phi1_e2
diag = integrate(f,(x,xl,xm)) + integrate(g,(x,xm,xr))
print("M_diagonal: ",simplify(diag))

f = phi1_e1*phi0_e1
sdiag = integrate(f,(x,xl,xm))
print("M_subdiag above: ",simplify(sdiag))

f = phi1_e2*phi2_e2
sdiag = integrate(f,(x,xm,xr))
print("M_subdiag below: ",simplify(sdiag))
```
